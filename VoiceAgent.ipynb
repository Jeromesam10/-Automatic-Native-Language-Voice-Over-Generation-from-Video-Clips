```python
%%writefile /content/tools/translator.py

from deep_translator import GoogleTranslator


def translate_text(

    text,

    target

):

    translated=GoogleTranslator(

        source='auto',

        target=target

    ).translate(text)


    return translated
```

```python
%%writefile /content/tools/speech.py

import edge_tts

async def generate_audio(

    text,

    voice,

    output

):

    communicate = edge_tts.Communicate(

        text,

        voice

    )

    await communicate.save(output)

    return output
```

```python
%%writefile /content/voice_agent.py

from tools.translator import translate_text
from tools.speech import generate_audio


class VoiceAgent:

    async def run(
        self,
        text,
        target_language,
        voice_sample=None
    ):

        language_map = {

            "tamil":"ta",
            "hindi":"hi",
            "telugu":"te",
            "malayalam":"ml",
            "kannada":"kn",
            "english":"en",
            "french":"fr",
            "german":"de",
            "spanish":"es",
            "japanese":"ja",
            "korean":"ko",
            "arabic":"ar",
            "russian":"ru",
            "chinese":"zh-cn",
            "portuguese":"pt",
            "italian":"it"

        }


        voice_map = {

            "ta":"ta-IN-PallaviNeural",

            "hi":"hi-IN-SwaraNeural",

            "te":"te-IN-ShrutiNeural",

            "ml":"ml-IN-SobhanaNeural",

            "kn":"kn-IN-SapnaNeural",

            "en":"en-US-JennyNeural",

            "fr":"fr-FR-DeniseNeural",

            "de":"de-DE-KatjaNeural",

            "es":"es-ES-ElviraNeural",

            "ja":"ja-JP-NanamiNeural",

            "ko":"ko-KR-SunHiNeural",

            "ar":"ar-SA-ZariyahNeural",

            "ru":"ru-RU-SvetlanaNeural",

            "zh-cn":"zh-CN-XiaoxiaoNeural",

            "pt":"pt-BR-FranciscaNeural",

            "it":"it-IT-ElsaNeural"

        }


        target_language = target_language.lower()


        lang_code = language_map.get(target_language)

        if lang_code is None:

           print("\nLanguage not supported")

           print("\nSupported Languages:\n")

        for lang in language_map:

            print(lang)

        return None
        if lang_code is None:

            print("Supported languages are:")

            print(list(language_map.keys()))

        return None


        translated = translate_text(

            text,

            lang_code

        )


        print("\nTranslated Text:\n")

        print(translated)


        voice = voice_map[lang_code]


        output="/content/outputs/output.mp3"


        await generate_audio(

            translated,

            voice,

            output

        )


        print("\nAudio Saved At:\n")

        print(output)


        return output
```
